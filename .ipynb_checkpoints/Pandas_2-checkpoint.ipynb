## Gözlem ve Değişken Seçimi: loc & iloc

```python
import numpy as np
import pandas as pd
```

```python
m = np.random.randint(1,30, size = (10,3))
df = pd.DataFrame(m, columns = ["var1","var2","var3"])
df
```

```python
# loc: tanımlandıgı şekli ile seçim yapmak için kullanılır 
```

```python
df.loc[0:3]                   # endpoint dahil ÖNEMLİ!!!!
```

```python
# iloc: alışık oldğumuz indexleme mantığı ile seçim yapar
```

```python
df.iloc[0:3]
```

```python
df.iloc[:3,:2]
```

```python
df.loc[0:3,"var3"]      # verilen kurallara bağlı olarak seçim yapılma ihtiyacı varsa gözlem yada 
                        # değişken isimlendirmeleri açısından loc kullan
```

```python
df.İloc[0:3,"var3"]     # loc ile iloc'un en belirgin farkı, yukarıdaki işlem iloc ile yapılınca hata verdi
```

```python
df.iloc[0:3]["var3"]     # verilen isimlendirmelerden bağımsız klasik index mantığı ile seçmek istersek iloc kullanılır
                         # hatanın önüne geçmek üzere değişken ayrı köşeli parantez ile oluşturuldu
```

```python
type(df["var1"])            # tek köşeli parantez ve çift köşeli parantez farkı
```

```python
type(df[["var1"]])
```

## Koşullu Eleman İşlemleri

```python
df
```

```python
df[0:2]
```

```python
df["var1"]                    # df["var1"] ile df.var1 aynı sonucu verir
```

```python
df["var1"][0:2]
```

```python
df[0:2][["var1","var2"]]   # ["var1","var2"] tek parantez olsaydı hata verirdi...fancy mantığı ile çift parantez olmalı
```

```python
df[df.var1 > 15]            # koşullu arama
```

```python
df[df.var1 > 15]["var1"]
```

```python
df[(df.var1 > 11) & (df.var3 < 8)]               # birden fazla koşul
```

```python
df.loc[(df.var1 > 15), ["var1","var2"]]           # (1) aşağıdaki ile iki farklı gösterim
```

```python
df[(df.var1 > 15)] [["var1","var2"]]               # (2) fancy ile 
```

```python
df[10 < df['var1'] < 15] # Bu şekilde sorgu yapamıyoruz. Bunun için .between() methodu var.
```

```python
df[df['var1'].between(10,15)]         # Default ayarlarda inclusive = 'both' dur. Yani iki taraf da dahildir.
```

```python
df[df['var1'].between(10,15 , inclusive = 'left')]       # 15'ler dahil olmadı
```

```python
df[~df['var1'].between(10,15)]         # " ~ " ile aralığın dışında kalanları getirir
```

```python
df.nlargest(5 , 'var2')                      # var2 için en büyük ilk 5 satırı getirdi
```

```python
# bu da aynı sonucu verir ama tercihimiz yukarıdaki özel fonksiyon olmalı 
df.sort_values('var2', ascending= False).head(5)    
```

```python
df.nsmallest(columns= ['var2'] , n = 3)       # var2 için en küçük ilk 3 satırı getirdi
```

```python
# Eğer parantez için sıra yanlış girildiyse hata almamak için o parametrenin ismiyle yazılmalı (n = 3 gibi)
```

```python
df.nsmallest(columns= ['var2','var3'] , n = 3)     # var2'yi baz alarak getirdi
```

## Birleştirme İşlemleri

```python
m = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(m, columns = ["var1","var2","var3"])
df1
```

```python
df2 = df1 +99
df2
```

```python
pd.concat([df1,df2])       # iki kere 0 dan 4'e index yazdı yani index problemi var
```

```python
pd.concat([df1,df2], ignore_index = True)      # index problemi ortadan kalktı ignore_index ile
```

```python
df1.columns
```

```python
df2.columns = ["var1","var2","deg3"]
df2
```

```python
pd.concat([df1,df2])          # düzgün bir şekilde birleşme olmaz problemli çünkü değişken farklılığı var
```

```python
pd.concat([df1,df2], join = "inner")    # iki kümenin kesişimlerini alır
```

## İleri Birleştirme İşlemleri

```python
df1 = pd.DataFrame({'calisanlar':['ali','veli','ayse','fatma'],
                    'grup':['muhasebe','muhendislik','muhendislik','ik']})
df1
```

```python
df2 = pd.DataFrame({'calisanlar':['ali','veli','ayse','fatma'],
                     'ilk_giris':[2010,2009,2014,2019]})
df2
```

```python
pd.merge(df1,df2)                     # otomatik ortak kısmı belirleyip birleştirir
```

```python
pd.merge(df1,df2, on = "calisanlar")   # calışanlar üstünden birleştir demek
```

```python
# coktan teke 
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.DataFrame({'grup':['muhasebe','muhendislik','ik'],
                    'mudur':['caner','mustafa','berkcan']})
df4
```

```python
pd.merge(df3,df4)
```

```python
# çoktan çoka
```

```python
df5 = pd.DataFrame({'grup':['muhasebe','muhasebe','muhendislik','muhendislik','ik','ik'],
                    'yetenekler':['matematik','excel','kodlama','linux','excel','yonetim']})
df5
```

```python
pd.merge(df1,df5)           # calışanlar çokladı
```

### Gruplama işlemleri

```python
df = pd.DataFrame({'gruplar':['A','B','C','A','B','C'],
                   'veri':[10,11,52,23,43,55]})
df
```

```python
df.groupby("gruplar")   # dataframe git, groupby fonk. ile gruplar değişkenini yakala ve kategori sınıflarını tut demek
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").sum()
```

```python
df = sns.load_dataset("planets")
df.head()
```

```python
df.groupby("method")                                # gruplama yapacak olduğu değişken method
```

```python
df.groupby("method")[["orbital_period"]].mean()      # gruplama neticesinde ilave bir değişken seçimi
```

```python
df[["method","orbital_period"]].groupby("method").mean()     # groupby 2. yöntem 
```

```python
df[["method","orbital_period"]].groupby("method").mean().sort_values(by = "orbital_period")   # sıralama
```

```python
df[["method","orbital_period"]].groupby("method", as_index = False).mean().sort_values(by = "orbital_period")
# index ilave etme
```

```python
df[["method","orbital_period","number"]].groupby(["method","number"]).mean()
```

```python
df.groupby("method")["orbital_period"].describe()   

# dataframe içerisindeki method kategorik değişkeninin sınıflarını yakaladı daha sonra 
# bu sınıfların altında orbital_period değişkennin her bir sınıf altındaki birçok değere
#(min, max, std vb.) ulaşmayı sağladı                                                     
```
